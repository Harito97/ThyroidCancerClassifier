## ver2

Input: data/processed/ver1

Các phép chuyển đổi:

- Với mỗi ảnh trong mỗi tập thực hiện detect tế bào, phân cụm để chọn vùng quan trọng và cắt ra đúng ảnh 4 ảnh size cắt là 224x224 (đại diện cho vùng quan trọng của ảnh) (mỗi ảnh đều phải có ảnh được cắt ra) (nếu số tế bào được detect trong 1 ảnh $\leq$ 1 threshold ta chia ảnh gốc thành 4 phần của 4 góc và resize 4 phần đó kích thước ảnh cắt)

- Với 4 ảnh được chọn cắt ra và 1 ảnh gốc resize về kích thước ảnh cắt $\to$ có 5 ảnh nên dễ vote

- Với mỗi ảnh được cắt ra đặt tên theo template 
```python
{ClassType}_{ImageID}_{CropID}.jpg # (CropID start từ 0: ảnh origin resize, 1..4 cho các ảnh được cắt)
```
Chú ý ```{ClassType}_{ImageID}_{0}.jpg``` sẽ chỉ là giả tưởng và không lưu lại thực tế để save bộ nhớ, khi load vào train model thì đọc trực tiếp từ data/processed/ver1 rồi resize về kích thước ảnh cắt
